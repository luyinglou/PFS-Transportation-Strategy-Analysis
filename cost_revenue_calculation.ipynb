```python
##calculate profit for each week
profit_2018=[]
profit_2019=[]
profit_2020=[]
for i in range(len(IB_volumne)):
    d_w = list(IB_volumne.iloc[i,:]/20000)
    d_w = [math.ceil(m) for m in d_w]
    ib_w_totalcost2018=(np.array(truckload_cost_ib_2018)*np.array(d_w)).sum()
    ib_w_totalcost2019=(np.array(truckload_cost_ib_2019)*np.array(d_w)).sum()
    ib_w_totalcost2020=(np.array(truckload_cost_ib_2020)*np.array(d_w)).sum()
    
    b_w = list(OB_volumne.iloc[i,:]/20000)
    ob_w_totalcost2018=(np.array(truckload_cost_ob_2018)*np.array(b_w)).sum()
    ob_w_totalcost2019=(np.array(truckload_cost_ob_2019)*np.array(b_w)).sum()
    ob_w_totalcost2020=(np.array(truckload_cost_ob_2020)*np.array(b_w)).sum()
    
    ibob_cost_w_2018=ib_w_totalcost2018+ob_w_totalcost2018
    ibob_cost_w_2020=ib_w_totalcost2020+ob_w_totalcost2020
    ibob_cost_w_2019=ib_w_totalcost2019+ob_w_totalcost2019

    profit_w_2020=(revenue['Total_Cost'])[i]-ibob_cost_w_2020
    profit_w_2019=(revenue['Total_Cost'])[i]-ibob_cost_w_2019
    profit_w_2018=(revenue['Total_Cost'])[i]-ibob_cost_w_2018
    profit_2018.append(profit_w_2018)
    profit_2019.append(profit_w_2019)
    profit_2020.append(profit_w_2020)
```

## To make profits

Findings:
1. If NET takes back all available orders, most of time NET could not make profits.
2. By observing the weekly profits calculated on IBOB cost in different years, we can find that profits decreased. This is becasue IBOB costs increased during 2018-2020, while all available orders didn't change too much.

Nextstep:
To make profits, NET cannot take back all available orders. NET can decide to not ship some orders since the volumes cannot support transportation costs. NET can also exclude some stations that have high lane costs. Try different combination of stations so that NET can grab as much volumnes as possible from other carriers.

## Conclusions

NET can make profits by excluding orders whose origin airport is among 'ABQ', 'BOI', 'DEN', 'GEG', 'LAS', 'PDX', 'PHX','SAN','SEA','SFO','SMF', because their volumes are not enough to support the shippment costs. This means NET can take back about 14 percent volumnes from other carriers.

After calculating the profits, NET can make positive profits in more than 90 percent of the time. The profits are calculated based on IBOB costs in 2020.

probability of making profits = number of weeks that have positive profits / total number of weeks
